- when you have >40% missing values in a column, you can simply drop the whole column unless that column is very important for any type of analysis
- when you have <40% missing values then you can handle those missing values
- when you have 2 or 3 missing values in the whole column then we can simply remove those rows

There are two cases to handle missing values - 
- numeric data column - we always replace the missing values with the average value of the column
- categorical data column - we always replace those missing values with the most frequent value of that column

```python
df = df.dropna(subset=['embarked'])
df.isnull().sum()
```

```python
avg_age = round(df['age'].mean())
print(avg_age)
```

```python
df['age'] = df['age'].fillna(avg_age)
df.isnull().sum()
```
