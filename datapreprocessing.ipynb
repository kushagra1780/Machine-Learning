````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer #It fills the missing value and calculates mean value of the missing values
from sklearn.pipeline import Pipeline
````

````python
penguins = sns.load_dataset("penguins")
penguins
````

````python
penguins.isnull().sum()
````

### Separating of numerical and categorical columns ###

````python
num_cols = penguins.select_dtypes('number').columns
print(num_cols.tolist())   #num_cols = numerical columns
````

````python
cat_cols = penguins.select_dtypes(exclude = 'number').columns
print(cat_cols.tolist())   #cat_cols = categorical columns
````

`Imputation`

````python
# initializing imputers for numerical and categorical columns

num_imp=SimpleImputer()# by default mean strategy is used ...likhna na likhna 
cal_imp=SimpleImputer(strategy='most_frequent')
````

````python
penguins[num_cols]=num_imp.fit_transform(penguins[num_cols])
penguins
````

````python
penguins.isnull().sum()
````

````python
penguins[cat_cols]=cal_imp.fit_transform(penguins[cat_cols])
penguins
````

````python
penguins.isnull().sum()
````

`ENCODING`

Encoding Categorical values

- ONE HOT ENCODING

````python
# features ke liye onehotencoder and labelencoder ka use target ke liye labelence
sex_enc = OneHotEncoder(drop = 'first')
sex_dummy = sex_enc.fit_transform(penguins[['sex']]).toarray()
sex_dummy_df = pd.DataFrame(sex_dummy)
````

drop first, to avoid ```dummy  variable``` trap......
- jab hum onehotencoding krte h to har category ke liye ek new column ban jata h
- agar aap n categories ko input kro to n columns banenge
- lekin inme se ek column humesha baki columns ke combination se predict kiya jaa skta h
- ise kehte h dummy variable trap aur ye multi collinearity create krta h
- jo linear models(jaise experience in job badhega to salary badhegi) ke liye problem ho skti h


```Multicollinearity```
- when more than one independent variable is strongly correlated with another independent variable

````python
sex_dummy_df #strongly correlated - multi-collinearity ...avoid krne hota h ....to avoid multi colinearity we drop one column by using drop = 'first' in OneHotEncoder
````

````python
cat_cols = ['species','island']
cat_enc = OneHotEncoder(drop = 'first')
dummy_cols = cat_enc.fit_transform(penguins[cat_cols]).toarray()
dummy_df = pd.DataFrame(dummy_cols)
````

````python
penguins['species'].value_counts()
````

````python
penguins['island'].value_counts()
````

````python
dummy_df
````

````python
clean_df = pd.concat([penguins, dummy_df, sex_dummy_df], 
                      axis=1).drop(columns = cat_cols +['sex'])
````
