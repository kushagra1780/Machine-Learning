`BENEFITS`

- Each machine starts with a random chance of being the best
- Every time you play, you update what you know about each machine
- Instead of just picking the machine with the highest average, you randomly choose based on how confident you are that a machine is the best

`Why is this better:`- Even if a machine was lucky early, Thompson Sampling still tests other machines in a smart way. It doesn't get stuck on a bad choice

```python
import random
```

```python
n_machines = 3
n_trials = 1000
```

```python
true_rewards = [0.2, 0.5, 0.8]
```

```python
success = [0] = n_machines
failures = [0] = n_machines
```

```python
# ==========================================================
# THOMPSON SAMPLING (MULTI-ARMED BANDIT)
# Beginner Friendly Version
# ==========================================================

# Random module import kar rahe hain
# Iska use random numbers generate karne ke liye hoga
import random


# ----------------------------------------------------------
# STEP 1 : Number of Slot Machines
# ----------------------------------------------------------

# Total kitni machines available hain
n_machines = 3


# ----------------------------------------------------------
# STEP 2 : Total Trials
# ----------------------------------------------------------

# Agent kitni baar machine choose karega
# Jitne zyada trials honge utna better learning hogi
n_trials = 1000


# ----------------------------------------------------------
# STEP 3 : Actual Winning Probability
# ----------------------------------------------------------

# Ye sirf Environment ko pata hai
# Agent ko ye values nahi pata

# Machine 1 -> 20% chance of winning
# Machine 2 -> 50% chance of winning
# Machine 3 -> 80% chance of winning

true_rewards = [0.2, 0.5, 0.8]


# ----------------------------------------------------------
# STEP 4 : Success Counter
# ----------------------------------------------------------

# Har machine kitni baar jeeti
# Initially kisi machine ne kuch nahi jeeta

success = [0] * n_machines


# ----------------------------------------------------------
# STEP 5 : Failure Counter
# ----------------------------------------------------------

# Har machine kitni baar haari

failures = [0] * n_machines


# ==========================================================
# MAIN LEARNING LOOP
# ==========================================================

for trial in range(n_trials):

    # ----------------------------------------------
    # STEP 6 : Sample Probability from Beta Distribution
    # ----------------------------------------------

    # Har machine ke liye ek probability sample hogi
    # Beta(success+1, failure+1)

    sampled_values = [

        random.betavariate(
            success[machine] + 1,
            failures[machine] + 1
        )

        for machine in range(n_machines)

    ]


    # ----------------------------------------------
    # STEP 7 : Choose Machine
    # ----------------------------------------------

    # Jis machine ka sampled probability
    # sabse zyada hai usko choose karo

    chosen_machine = sampled_values.index(max(sampled_values))


    # ----------------------------------------------
    # STEP 8 : Check Reward
    # ----------------------------------------------

    # Environment reward generate karega

    # Example

    # Machine3 probability = 0.8

    # Agar random number 0.45 hai
    # Reward milega

    # Agar random number 0.91 hai
    # Reward nahi milega

    reward = 1 if random.random() < true_rewards[chosen_machine] else 0


    # ----------------------------------------------
    # STEP 9 : Update Knowledge
    # ----------------------------------------------

    # Agar reward mila

    if reward:

        # Success count increase
        success[chosen_machine] += 1

    else:

        # Failure count increase
        failures[chosen_machine] += 1


# ==========================================================
# FINAL RESULTS
# ==========================================================

print("\n========= FINAL RESULT =========\n")

for machine in range(n_machines):

    # Total kitni baar machine kheli gayi

    total_played = success[machine] + failures[machine]


    # Agar machine kabhi choose hi nahi hui

    if total_played == 0:

        print(f"Machine {machine+1}")
        print("Never Played\n")

    else:

        # Estimated Win Rate

        estimated_rate = success[machine] / total_played

        print(f"Machine {machine+1}")
        print(f"Played          : {total_played} times")
        print(f"Wins            : {success[machine]}")
        print(f"Losses          : {failures[machine]}")
        print(f"Estimated Rate  : {estimated_rate:.2f}")
        print("-" * 40)


# ==========================================================
# BEST MACHINE
# ==========================================================

# Jis machine ke sabse zyada wins hain

best_machine = success.index(max(success))

print(f"\nBest Machine to Play : Machine {best_machine + 1}")
```
